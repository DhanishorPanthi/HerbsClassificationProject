# All Data

```python
import os
import matplotlib.pyplot as plt


BASE_DIR = "D:\Ashutosh\Herbs\Cleaned_Data\All"
TRAIN_DIR = os.path.join(BASE_DIR, "Train")
VAL_DIR = os.path.join(BASE_DIR, "Val")

# Path to your folder
folder_path = BASE_DIR

# List of valid image file extensions
valid_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff'}

# Dictionary to store class names and their sample counts
class_counts = {}

# Function to count valid sample files in a directory
def count_valid_samples(directory):
    count = 0
    for filename in os.listdir(directory):
        # Skip hidden files or directories starting with '._'
        if filename.startswith('._'):
            continue
        if os.path.isfile(os.path.join(directory, filename)):
            if os.path.splitext(filename)[1].lower() in valid_extensions:
                count += 1
    return count

# Iterate through each class folder and count the number of valid sample files
for class_name in os.listdir(folder_path):
    class_folder = os.path.join(folder_path, class_name)
    if os.path.isdir(class_folder):
        sample_count = count_valid_samples(class_folder)
        class_counts[class_name] = sample_count

# Sort the class_counts dictionary by number of samples in ascending order
sorted_class_counts = dict(sorted(class_counts.items(), key=lambda item: item[1]))

# Get the number of classes
number_of_classes = len(class_counts)

# Print the number of classes
print(f"Number of classes: {number_of_classes}")

# Generate labels with class number and name
class_labels = [f"{i+1}. {name}" for i, name in enumerate(sorted_class_counts.keys())]

# Plotting the data
plt.figure(figsize=(20, 10))
plt.bar(class_labels, sorted_class_counts.values())
plt.axhline(y=1000, color='blue', linestyle='--', linewidth=2)  # Add horizontal line at y = 1000
plt.axhline(y=700, color='red', linestyle='--', linewidth=2)  # Add horizontal line at y = 800
plt.xlabel('Class Name')
plt.ylabel('Number of Samples')
plt.title(f'Number of Samples in Each Class (Total Classes: {number_of_classes})')
plt.xticks(rotation=90)  # Rotate class names for better visibility
plt.tight_layout()  # Adjust layout to make space for labels
plt.show()
```
